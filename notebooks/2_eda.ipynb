```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## 1. Load the Data

```python
df=pd.read_parquet('../data/wrangled_df.parquet')
```

## 2. EDA

### 2.1. Explore the Data

```python
# Look at histograms for each feature
df.hist(bins=15, figsize=(15, 10), layout=(3, 3))
plt.tight_layout()
plt.show()
```

- Local Event Date: Clearly trends down over time. Take a closer look with a line graph.
- Total Fatalities: Low numbers of fatalities are most common by a long shot. Indicative of light planes (low capacity) experiencing the most accidents?
- Total Injuries: Similar results to total fatalities. Explore both more to see how often higher numbers occur.
- Nbr of Engines: Primarily single and twin engine planes. Again, indicative of smaller aircraft.
- PIC Flight Time Total Hrs: Clear spike in accidents with less experienced pilots.
- PIC Flight Time Total Make-Model: Similar to total hrs. Explore both in more detail.

```python
# Bar Plot: Aircraft Damage

# Get count
count=df['Aircraft Damage'].value_counts(ascending=False)

# Set the size of the plot
plt.figure(figsize=(7, 5))

# Plot the bar plot
count.plot(kind='bar')

# Set title and labels
plt.xlabel('Aircraft Damage Severity')
plt.ylabel('Count')

# Rotate x-axis labels for better readability and increase font size
plt.xticks(rotation=90, fontsize=8)

# Show plot
plt.show()
```

```python
df['Aircraft Damage'].value_counts()
```

The vast marjority of damage is minor, with very few incidents of substantial damage and extremely few instances of destroyed aircraft. 

```python
# Bar Plot: Flight Phase

# Get count
count=df['Flight Phase'].value_counts(ascending=False)

# Set the size of the plot
plt.figure(figsize=(15, 8))

# Plot the bar plot
count.plot(kind='bar')

# Set title and labels
plt.xlabel('Flight Phase')
plt.ylabel('Count')

# Rotate x-axis labels for better readability and increase font size
plt.xticks(rotation=90, fontsize=10)

# Show plot
plt.show()
```

- Level off touchdown: the point of transition from nose-low to nose-up attitude, just before landing, until touchdown
- Roll-out (fixed wing): the final phase of landing when focus shifts from touchdown to taxiing

```python
# Create year column
df['Year']=pd.to_datetime(df['Local Event Date'], format='%Y').dt.year

# Convert to int to group into bins
df['Year']=df['Year'].astype(int)
```

```python
# Group Year into bins
bins=[0,1985,1990,1995,2000,2005,2010,2015,2020,2025]
labels=['1985','1990','1995','2000','2005','2010','2015','2020','2025']
df['Date Bins'] = pd.cut(df['Year'], bins=bins, labels=labels)

#df.head()
```

```python
df['Date Bins'].value_counts()
```

```python
# Add Year to new df
year_df=df[['Date Bins']] #, 'incident_type'

# Add counts column 
year_df['Count']=year_df.groupby(['Date Bins'])['Date Bins'].transform('count') #, 'incident_type'

year_df.set_index('Date Bins')

year_df.head()
```

```python
# Line Plot: Year and Accident Count

# Adjust plot size
plt.figure(figsize=(8, 6))

# Plot Year and Count
p=sns.lineplot(data=year_df, x='Date Bins', y='Count') #, hue='incident_type'

# Adjust legend 
#sns.move_legend(p, 'lower center', bbox_to_anchor=(.5, 1), ncol=5, title=None, frameon=False)
```

This very clearly shows that the number of accidents has drastically decreased over time.
